```python
import matplotlib.pyplot as plt

# Trade systems and consistent inflation-adjusted values
systems = ['Mercantilism', 'Dutch Trade', 'Anglo-British Trade']
start_years = [1500, 1600, 1700]
end_years = [1750, 1700, 1900]
profit_margin = [6, 9, 8]
economic_growth = [5, 7, 9]
trade_value_usd = [8, 20, 45]  # All in billions USD (2025-adjusted)

fig, ax = plt.subplots(figsize=(12, 6))

# Draw timeline bars and annotations
for i in range(len(systems)):
    ax.barh(i, end_years[i] - start_years[i], left=start_years[i], height=0.6, alpha=0.7)
    
    ax.text(
        end_years[i] + 5, i,
        f'Profit: {profit_margin[i]}/10\n'
        f'Growth: {economic_growth[i]}/10\n'
        f'Trade: ${trade_value_usd[i]}B/yr',
        va='center', fontsize=10
    )

# Format timeline
ax.set_yticks(range(len(systems)))
ax.set_yticklabels(systems)
ax.set_xlabel("Year")
ax.set_title("Trade Systems Timeline with Standardized Economic Metrics (2025 USD)")
ax.set_xlim(1450, 1950)
ax.grid(axis='x', linestyle='--', alpha=0.5)

# Add inflation reference box in top-right corner
plt.gcf().text(
    0.73, 0.75,
    " All trade values adjusted\n   to 2025 USD\ Avg inflation adj:\n   ~2.5%/yr over ~400 yrs",
    fontsize=9,
    bbox=dict(facecolor='lightyellow', edgecolor='gray')
)

plt.tight_layout()
plt.show()
```
