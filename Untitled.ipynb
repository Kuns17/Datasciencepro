```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Timeline data
systems = ['Mercantilism', 'Dutch Trade', 'Anglo-British Trade']
start_years = [1500, 1600, 1700]
end_years = [1750, 1700, 1900]
profit_margin = [6, 9, 8]
economic_growth = [5, 7, 9]

fig, ax = plt.subplots(figsize=(12, 6))

# Plot timeline bars
for i, system in enumerate(systems):
    ax.barh(i, end_years[i] - start_years[i], left=start_years[i], height=0.6, alpha=0.7)

# Add labels and scores
for i, system in enumerate(systems):
    ax.text(end_years[i] + 5, i, f'Profit: {profit_margin[i]}/10\nGrowth: {economic_growth[i]}/10', va='center')
# Formatting
ax.set_yticks(range(len(systems)))
ax.set_yticklabels(systems)
ax.set_xlabel("Year")
ax.set_title("Timeline of Trade Systems with Profit & Growth Scores (Historical Estimate)")
ax.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
