```python
import matplotlib.pyplot as plt
# Data
systems = ['Dutch Trade', 'Anglo-British Trade', 'Mercantilism']
start_years = [1600, 1700, 1550]
end_years = [1750, 1900, 1800]
currencies = ['Guilder', 'GBP', 'Portuguese Escudo']
historical_values = [2.5, 4.0, 1.8]  # in billions (original currencies)
conversion_to_usd = [0.60, 1.25, 0.50]
inflation_multipliers = [35, 36, 35]
economic_growth = [8.2, 9.1, 6.5]

# USD 2025 calculation
usd_values_2025 = []
for val, conv, infl in zip(historical_values, conversion_to_usd, inflation_multipliers):
    current_val = val * conv * infl
    usd_values_2025.append(round(current_val, 2))

fig, ax = plt.subplots(figsize=(14, 6))
ax.set_xlim(1500, 1950)
ax.set_ylim(-1, len(systems))

# Plot bars and labels
for i in range(len(systems)):
    ax.barh(i, end_years[i] - start_years[i], left=start_years[i], height=0.4, color='steelblue')
    label = (
        f"Trade: {historical_values[i]}B {currencies[i]}\n"
        f"Now: ${usd_values_2025[i]}B USD\n"
        f"Economic Growth: {economic_growth[i]}/10"
    )
    ax.text(end_years[i] + 5, i, label, va='center', fontsize=10)

# Currency conversion table (left)
conversion_text = (
    "Currency Conversion Rates:\n"
    "1 Guilder  $0.60\n"
    "1 GBP  $1.25\n"
    "1 Escudo  $0.50"
)
plt.text(1510, 2.5, conversion_text, fontsize=9,
         bbox=dict(boxstyle='round,pad=0.5', facecolor='lightyellow', edgecolor='black'))

# Axes and title
ax.set_yticks(range(len(systems)))
ax.set_yticklabels(systems)
ax.set_xlabel('Year')
ax.set_title('Historical Trade Systems ', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Historical trade data (simplified and converted to USD billions)
data = {
    "System": ["Mercantilism", "Dutch East India Co.", "British East India Co."],
    "Start Year": [1500, 1602, 1600],
    "End Year": [1800, 1799, 1874],
    "Trade Volume (USD Billion)": [500, 7.9, 18.0],  # Adjusted for scale
    "Economic Growth (%)": [2.5, 5.0, 6.5],
}

df = pd.DataFrame(data)

# Plot setup
fig, ax1 = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = range(len(df))

# Plot Trade Volume
ax1.bar(index, df["Trade Volume (USD Billion)"], bar_width, label="Trade Volume (USD B)", color='skyblue')
ax1.set_ylabel("Trade Volume (USD Billion)")
ax1.set_title("Historical Trade Comparison of Mercantilism, Dutch & British EIC")
ax1.set_xticks(index)
ax1.set_xticklabels(df["System"])

# Create second y-axis for Economic Growth
ax2 = ax1.twinx()
ax2.plot(index, df["Economic Growth (%)"], color='green', marker='o', linewidth=2, label="Economic Growth (%)")
ax2.set_ylabel("Estimated Economic Growth (%)")

# Legends
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.92))
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
